```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load the model and tokenizer
model_name = "tayyibsupercool/Phi_3.5_mini-resource_allocation-spectral_efficiecy_rician_k_12_instruct_10k"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Ensure the model is in evaluation mode
model.eval()

# Define the function to chat with the model
def chat_with_model(prompt, max_length=100):
    # Encode the prompt to tokens
    inputs = tokenizer(prompt, return_tensors="pt")
    
    # Generate a response from the model
    with torch.no_grad():
        outputs = model.generate(inputs["input_ids"], max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2)
    
    # Decode the generated response back to text
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# Test the model with a prompt
prompt = "Hello, how can I assist you with resource allocation?"
response = chat_with_model(prompt)
print(response)
```
